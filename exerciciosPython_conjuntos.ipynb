```python
# ----------------------------------- #

A = {1, 2, 3}
B = {3, 4, 5}

print("União:", A | B)
```

**Exercício 15 – Desafio**
- Crie um programa que:
1. Defina dois conjuntos A e B.
2. Mostre:

*   União
*   Interseção
*   Diferença A − B
*   Diferença B − A
*   Diferença simétrica

```python
# ----------------------------------- #

A = {1, 2, 3, 4}
B = {3, 4, 5, 6}

print("União:", A | B)
print("Interseção:", A & B)
print("A - B:", A - B)
print("B - A:", B - A)
print("Diferença simétrica:", A ^ B)
```
